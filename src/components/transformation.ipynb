```python
#Fonction pour remplacer les noms de colonnes
def replace_col_names(df, unite):
    new_cols = []
    ue_count = {}  # Dictionnaire pour compter le nombre de colonnes par UE

    for col in df.columns:
        found = False
        for ue, col_names in unite.items():
            if col in col_names:
                # Compter combien de fois cette UE a déjà été utilisée
                ue_count[ue] = ue_count.get(ue, 0) + 1
                # Ajouter un suffixe numérique si nécessaire
                new_col = f"{ue}_{ue_count[ue]}" if ue_count[ue] > 1 else ue
                new_cols.append(new_col)
                found = True
                break
        if not found:
            new_cols.append(col)  # Conserver le nom d'origine si la colonne n'est pas dans le dictionnaire
    df.columns = new_cols

    return df
```

```python
concatenated_df = pd.DataFrame()

for file in list_path:
    if file.endswith((".xlsx", ".xls")):
        file_path = os.path.join(path, file)
        df = pd.read_excel(file_path).dropna(axis=0, how='all')

        # Appliquer la fonction pour renommer les colonnes selon les catégories
        df = replace_col_names(df, unite)

        # Conserver uniquement les colonnes des nouvelles catégories + colonnes d'identité
        categories = list(unite.keys())
        colonnes_a_garder = ['N°', 'Prénom(s)', 'Nom'] + categories
        df = df.loc[:, df.columns.intersection(colonnes_a_garder)]

        # Extraire l’année et le semestre depuis le nom du fichier
        annee = file.split("_")[0]
        semestre = file.split("_")[2]
        df["Année"] = annee
        df["Semestre"] = semestre

        concatenated_df = pd.concat([concatenated_df, df], ignore_index=True)
concatenated_df.fillna(0, inplace=True)


# Aperçu du DataFrame final
concatenated_df.head()
```

```python
semester_mapping = {
    '20-21': {'S1': 'S1', 'S2': 'S2'},
    '21-22': {'S1': 'S3', 'S2': 'S4'},
    '22-23': {'S1': 'S5', 'S2': 'S6'},
}

# Fonction pour mettre a jour la colonne semester
def update_semester(row):
    academic_year = row['Année']
    semester = row['Semestre']
    if academic_year in semester_mapping and semester in semester_mapping[academic_year]:
        return semester_mapping[academic_year][semester]
    return semester

# Appliquer la fonction en ajoutant un nouveau colonne
concatenated_df['Updated_Semestre'] = concatenated_df.apply(update_semester, axis=1)

concatenated_df['Semestre'] = concatenated_df['Updated_Semestre']
concatenated_df.drop('Updated_Semestre', axis=1, inplace=True)

concatenated_df.head(10)
```

```python
# Fonction pour calculer l'âge
def calculate_age(date_naissance, annee):
    try:
        # Extraire la dernière année de la plage (par exemple, '21' dans '20-21')
        annee_split = annee.split('-')
        last_annee = int(annee_split[-1])

        # Ajouter 2000 pour obtenir l'année complète (par exemple, 2021)
        full_year = 2000 + last_annee
    except (ValueError, AttributeError, IndexError):
        # Gérer les cas où 'annee' n'est pas au format attendu
        return np.nan

    # Vérifier si 'date_naissance' est valide
    if pd.isna(date_naissance):
        return np.nan

    # Calculer l'âge
    age = full_year - date_naissance.year
    return age

# Appliquer la fonction pour créer une nouvelle colonne 'age'
df['age'] = df.apply(lambda row: calculate_age(row['date_naissance'], row['Année']), axis=1)

# Afficher le DataFrame avec la nouvelle colonne 'age'
df.head()
```
